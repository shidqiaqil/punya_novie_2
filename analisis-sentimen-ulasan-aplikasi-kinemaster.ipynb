# **Pre Processing**

**Case Folding**

```python
#my_df = variabel untuk menyimpan data asli
#my_df_prep = Variabel untuk menyimpan data hasil pre processing
my_df_prep = my_df
```

```python
#Proses case folding: proses mengubah seluruh huruf menjadi huruf kecil.
my_df_prep["content"] = my_df_prep["content"].str.lower()
```

```python
my_df_prep
```

**Tokenize**

```python
# Import the NLTK library
import nltk
```

```python
# Tokenize data
my_df_prep["tokens"] = my_df_prep["content"].apply(nltk.word_tokenize)
```

```python
#hasil tokenize data
my_df_prep
```

**Filtering Stemming**

```python
from nltk.corpus import stopwords
```

```python
#List kata stopwords (kata-kata umum yang tidak memiliki makna penting)
stopwords.words('indonesian')
```

```python
# Filter the tokens (stopwords). 
# Menghapus kata-kata umum yang tidak memiliki makna penting. Fungsi ini mengembalikan daftar kata-kata umum dalam bahasa Indonesia.
stopwords = nltk.corpus.stopwords.words("indonesian")
my_df_prep["filtered_tokens"] = my_df_prep["tokens"].apply(lambda x: [t for t in x if t not in stopwords])
```

```python
# Stemming the tokens
# Melakukan stemming pada token yang telah difilter. Fungsi ini akan mengubah kata ke bentuk dasarnya.
stemmer = nltk.stem.PorterStemmer()
my_df_prep["stemmed_tokens"] = my_df_prep["filtered_tokens"].apply(lambda x: [stemmer.stem(t) for t in x])
```

**Penjelasan setiap kolom**
1. **at**              = tanggal untuk setiap komentar
2. **content**         = komentar pengguna
3. **tokens**          = content yang telah di tokenize (membagi string menjadi daftar token,    di mana setiap token adalah kata atau tanda baca)
4. **filtered_tokens**  = Menghapus kata-kata umum yang tidak memiliki makna penting
5. **stemmed_tokens**   = mengubah kata ke bentuk dasarnya
